```python
import pandas as pd
```

```python
data = pd.read_csv("ecg.csv", header=None)
```

```python
data.head()
```

```python
nbre_rows = data.shape[0] #number of rows
```

```python
nbre_columns = data.shape[1] #number of features or columns
```

### we should provide a name for every feature

```python
columns = [f"feature_{i}" for i in range(1, nbre_columns)] + ["label"]
data.columns = columns
```

```python
data.to_csv("ecg2.csv", index=False)
```

```python
data.head()
```

```python
class_counts = data['label'].value_counts()
class_percentages = (class_counts / class_counts.sum()) * 100
class_percentages
```

```python
plt.figure(figsize=(6, 4))
class_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Class Distribution in Label Column')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.xticks([0, 1], ['Class 0', 'Class 1'], rotation=0)
plt.show()
```
